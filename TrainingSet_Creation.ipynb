```python
import pandas as pd
import numpy as np
```

```python
#df = pd.read_csv("Restaurant_Reviews.tsv",delimiter = '\t')
df = pd.read_csv("rest.csv")
#df = df.rename(columns={'Review': 'text'})
df = df.rename(columns={'text_': 'text'})
df['topic'] = 'restaurant'
restaurants = df[['text','topic']].copy()
restaurants
```

```python
df1 = pd.read_csv("Books_rating.csv")
```

```python
df1 = df1.rename(columns={'review/text': 'text'})
df1['topic'] = 'book'
books = df1[['text','topic']].copy()
books
```

```python
books = books.sample(n=1000)
books = books.reset_index()
books = books[['text','topic']].copy()
books
```

```python
df2 = pd.read_csv("IMDB.csv")
```

```python
import pathlib
import sklearn
import numpy
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from collections import Counter
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
final_vec = CountVectorizer(min_df=2, 
                             tokenizer=nltk.word_tokenize, 
                             stop_words=stopwords.words('english'))
```

```python
cutoff = len(train.text)
#cut = len(test.text)

#testing = (test.text.tolist()+extra.text.tolist())

review_counts = final_vec.fit_transform(train.text.tolist()+test.text.tolist())

review_trains_counts = review_counts[:cutoff]
review_test_counts = review_counts[cutoff:]

#review_test_counts = review_testing_counts[:cut]
#extra_counts = review_testing_counts[cut:]

print(review_trains_counts.shape)
print(review_test_counts.shape)
#print(extra_counts.shape)
```

```python
tfidf_transformer = TfidfTransformer()
review_trains_tfidf = tfidf_transformer.fit_transform(review_trains_counts)
review_trains_tfidf.shape
#(30000, 48352)
#(10, 48352)
```

```python
review_test_tfidf = tfidf_transformer.fit_transform(review_test_counts)
review_test_tfidf.shape
```

```python
from sklearn import svm
lin_clf = svm.LinearSVC()
lin_clf.fit(review_trains_tfidf, train.topic)
predict = lin_clf.predict(review_test_tfidf)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test.topic, predict))
```
